X,y split:

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_down.drop('Churn', axis=1)
y= df_down['Churn']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=123)
```

Model: Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
randfor = RandomForestClassifier(max_depth=10,max_features="sqrt",n_estimators=100,
                             bootstrap=True,oob_score=True, random_state=123)
randfor.fit(X_train, y_train)
print(randfor.score(X_train, y_train))
print(randfor.score(X_test, y_test))
```

```python
from sklearn.model_selection import cross_val_score
cross_val_scores = cross_val_score(randfor, X_train, y_train, cv=5)
print(cross_val_scores,
np.mean(cross_val_scores),
np.std(cross_val_scores))
```

Upsampling

```python
df = pd.read_csv('DATA_Customer-Churn_magda.csv')
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].mean())
numericalX = df.select_dtypes(np.number)
categoricalX = df.select_dtypes('object')
categoricalX  = pd.get_dummies(categoricalX, drop_first=True)
df = pd.concat([numericalX, categoricalX], axis = 1)
```

```python
df
```

```python
X = df.drop('Churn_Yes', axis=1)
y= df['Churn_Yes']
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X_sm, y_sm = smote.fit_resample(X, y)
y_sm.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.2, random_state=123)
```

```python
randfor_up = RandomForestClassifier(max_depth=10,max_features="sqrt",n_estimators=100,
                             bootstrap=True,oob_score=True, random_state=123)
randfor_up.fit(X_train, y_train)
print(randfor_up.score(X_train, y_train))
print(randfor_up.score(X_test, y_test))
```

```python
randfor_up.predict_proba(X_test) #predicting the pobability of 0 and 1 of each observation
```

GridSearch:

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [500, 600],
    'min_samples_split': [2, 4],
    'min_samples_leaf' : [1, 2],
    'max_features': ['sqrt']
    ##'max_samples' : ['None', 0.5],
    ##'max_depth':[3,5,10],
    ## 'bootstrap':[True,False] 
    }
```

```python
grid_search = GridSearchCV(randfor_up, param_grid, cv=5,return_train_score=True,n_jobs=-1,)
```

```python
grid_search.fit(X_train,y_train)
grid_search.best_params_ #To check the best set of parameters returned
```
